## 2. Описание постановки задачи

**Набор данных: Предпринимательская компетентность у студентов университета**

Набор данных состоит из 16 признаков и целевой переменной, предсказывающей, обладает ли студент предпринимательскими способностями или нет.
Целевая переменная дискретна и представляет собой 0 или 1 в зависимости от того, станет ли студент предпринимателем или нет.
К входным переменным относятся:
    
1. EducationSector (Сфера образования) - "Engineering Sciences", "Economic Sciences, Business Studies, Commerce and Law", "Art, Music or Design", "Medicine, Health Sciences" и другие
    
2. IndividualProject - Да/Нет в зависимости от наличия индивидуального проекта
    
3. Age - Возраст студента в годах
    
4. Gender - пол студента
    
5. City - Да/Нет в зависимости от того, живет ли студент в городе
    
6. Influenced - Да/Нет в зависимости от наличия чьего-либо влияния на студента
    
7. Perseverance - оценка настойчивости студента от 1 до 5
    
8. DesireToTakeInitiative  - оценка желания проявлять инициативу у студента от 1 до 5
    
9. Competitiveness - оценка конкурентноспособности от 1 до 5
    
10. SelfReliance - оценка самодостаточности от 1 до 5
    
11. StrongNeedToAchieve - оценка сильного желания достигнуть цели от 1 до 5
    
12. SelfConfidence - оценка уверенности в себе от 1 до 5
    
13. GoodPhysicalHealth - оценка физического здоровья от 1 до 5
    
14. MentalDisorder - Да/Нет в зависимости от наличия ментальных расстройств
    
15. KeyTraits - ключевые черты характера студента ("Positivity", "Passion" и т.д.)
    
16. ReasonsForLack - причина отсутствия желания предпринимательской деятельности (Null, "Not interested" и т.д.)
